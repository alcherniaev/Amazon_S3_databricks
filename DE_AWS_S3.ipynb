```
# file location and type
file_location = "dbfs:/mnt/s3data/listings.csv"
file_type = "csv"

# CV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. 
df_listing = spark.read.format(file_type) \
                  .option("infer_schema", infer_schema) \
                  .option("header", first_row_is_header) \
                  .option("sep", delimiter) \
                  .load(file_location)

display(df_listing)
```

```
df_listing = spark.read.format(file_type) \
                  .option("infer_schema", infer_schema) \
                  .option("header", first_row_is_header) \
                  .option("sep", delimiter) \
                  .option("multiline", "true") \
                  .load(file_location)

display(df_listing)
```

```
df_listing = spark.read.format(file_type) \
                  .option("infer_schema", infer_schema) \
                  .option("header", first_row_is_header) \
                  .option("sep", delimiter) \
                  .option("multiline", "true") \
                  .option("quote", "\"") \
                  .option("escape", "\"") \
                  .load(file_location)

display(df_listing)
```

```
%sql

drop table if exists listing
```

```
listing_table = "listing"
df_listing.write.format("parquet").saveAsTable(listing_table)
```
