```
from pyspark.sql.types import *

# file location and type
file_location = "dbfs:/mnt/s3data/neighbourhoods.csv"
file_type = "csv"

# CV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

neighbourhoodSchema = StructType([
  StructField("neighbourhood_group", StringType(), True),
  StructField("neighbourhood", StringType(), True),
])

# The applied options are for CSV files. 
df_neighbourhoods = spark.read.format(file_type) \
                  .option("schema", neighbourhoodSchema) \
                  .option("header", first_row_is_header) \
                  .option("sep", delimiter) \
                  .option("multiline", "true") \
                  .option("quote", "\"") \
                  .option("escape", "\"") \
                  .load(file_location)

display(df_neighbourhoods.limit(5))
```

```
listing_table = "neighbourhood"
df_neighbourhoods.write.mode("overwrite").format("parquet").saveAsTable(listing_table)
```

```
# file location and type
file_location = "dbfs:/mnt/s3data/reviews.csv"
file_type = "csv"

# CV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","


# The applied options are for CSV files. 
df_reviews = spark.read.format(file_type) \
                  .option("inferSchema", infer_schema) \
                  .option("header", first_row_is_header) \
                  .option("sep", delimiter) \
                  .option("wholeFile", "true") \
                  .option("multiline", 'true') \
                  .load(file_location)

display(df_reviews.limit(5))
```

```
from pyspark.sql.types import FloatType

def trim_char(string):
  return string.strip("$")

spark.udf.register('trim_func', trim_char) # for spark sql

trim_func_udf = udf(trim_char) # for DataFrame
```

```
df_avg_price = df_listing.select("property_type", "price", trim_func_udf("price").cast(FloatType()).alias("price_f")).groupBy("property_type").agg(F.round(F.avg(('price_f')),2).alias('avg_price'))
df_avg_price.show(truncate=False)
```

#### Percentage of listing on Neighbourhood ?

```
from pyspark.sql.window import Window

df_neigh_count = df_listing.groupBy("neighbourhood").agg(F.count("neighbourhood")).withColumnRenamed("count(neighbourhood)", "count").orderBy("count", ascending=False)
df_neigh_count.show(truncate=False, n=10)
```

```
window = Window.partitionBy()
df_neigh_count = df_neigh_count.withColumn("percentage", F.round(F.col("count") / F.sum("count").over(window) * 100, 3 )).orderBy("percentage", ascending=False)
display(df_neigh_count)
```

#### Find the customer who travels the most based on reviews data

```
df_reviews_fil.groupBy("reviewer_id").count().filter(F.col("count") >= 15).orderBy("count", ascending=False).show()
```

```
df_reviews_cnt = df_reviews_fil.groupBy("reviewer_id").count().filter(F.col("count") >= 15)
```

```
from datetime import datetime
dfunc = udf(lambda x: datetime.strptime(x, "%Y-%m-%d"), DateType())
spark.udf.register("datec", dfunc)
```

```
df_reviews_dt = df_reviews_fil.withColumn("datetime", dfunc(F.col("date"))).groupBy("reviewer_id").agg(F.min("datetime"), F.max("datetime"))
display(df_reviews_dt)
```

```
df_reviews_cnt.show(5)
```

```
df_reviews_summary = df_reviews_cnt.join(df_reviews_dt, "reviewer_id").withColumn("date_diff_days", F.datediff("max(datetime)", "min(datetime)")).orderBy("count", asceding=False)
```

```
display(df_reviews_summary)
```

```
df_reviews_summary.createOrReplaceTempView("review_summary")
```

```
%sql
with cte as (
              select max(date_diff_days) as max_date from review_summary
            )
            
select max_date, reviewer_id as most_traveled_customer_id from review_summary join cte on cte.max_date = review_summary.date_diff_days

```

```
%sql
select * from review_summary
```

```
%sql

select reviewer_id, round(date_diff_days / count , 2) as avg_days from review_summary order by avg_days asc
```

#### Rank the top 2 listing in each neighbourhood based on reviews

```
%sql

select distinct listing_id, neighbourhood, price, count(listing_id) over (partition by listing_id order by listing_id) as review_cnt 
from ( 
  select listing_id, neighbourhood, price, reviewer_id
  from listing l left join reviews_fixed r on l.id = r.listing_id
)
```
